```python
# Import required libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Create sample data
data = {
    'Product': ['A', 'B', 'C', 'D', 'E'],
    'January': [120, 90, 150, 130, 100],
    'February': [140, 110, 160, 115, 105],
    'Sales': [260, 200, 310, 245, 205],
    'Ratings': [4.1, 3.8, 4.5, 4.0, 3.9]
}

df = pd.DataFrame(data)
```

```python
plt.figure(figsize=(8, 5))
plt.plot(df['Product'], df['January'], label='January', marker='o')
plt.plot(df['Product'], df['February'], label='February', marker='s')
plt.title("Monthly Sales per Product")
plt.xlabel("Product")
plt.ylabel("Units Sold")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.bar(df['Product'], df['Sales'], color='skyblue')
plt.title("Total Sales by Product")
plt.xlabel("Product")
plt.ylabel("Units Sold")
plt.show()
```

```python
plt.figure(figsize=(6, 6))
plt.pie(df['Sales'], labels=df['Product'], autopct='%1.1f%%', startangle=90)
plt.title("Sales Distribution")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.hist(df['Ratings'], bins=5, color='green', edgecolor='black')
plt.title("Distribution of Product Ratings")
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(data=df[['January', 'February']], palette='Set2')
plt.title("Boxplot of Monthly Sales")
plt.ylabel("Units Sold")
plt.show()
```
